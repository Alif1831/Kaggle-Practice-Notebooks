```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk(''):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
fastfood= pd.read_csv("Top 50 Fast-Food Chains in USA.csv")
fastfood.set_index('Fast-Food Chains', inplace = True)
```

```python
fastfood.columns
```

```python
fastfood.head()
```

```python
fastfood["Proportion of franchise"] = round(fastfood["Franchised Stores"] / fastfood["2021 Total Units"], 2)
```

```python
fastfood["Proportion of franchise"].head()
```

```python
p = 0
for i in range(9):
    p += 1
    q = len(fastfood[fastfood["Proportion of franchise"]<(p/10)])
    print(f"{q} out of the 50 restaurants franchise less than {p*10} percent of their stores")
    
print("And 28 out of the 50 restaurants franchise more than 90 percent of their stores") 

```

```python
uswide = fastfood.sort_values("U.S. Systemwide Sales (Millions - U.S Dollars)", ascending = False)
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize = (16,6))

plt.title("U.S. Systemwide Sales (Millions - U.S Dollars)", fontsize = 16)
plt.xticks(rotation = 90)
plt.bar(uswide.index, uswide["U.S. Systemwide Sales (Millions - U.S Dollars)"], color = "#74b9ff", label = "Average Sales per unit")
plt.ylabel("Millions - U.S Dollars", fontsize = 12)
plt.legend()
```

```python
uswide[["U.S. Systemwide Sales (Millions - U.S Dollars)", "Proportion of franchise"]]
```

```python
plt.figure(figsize = (4,4))
plt.title("Relationship between U.S. Systemwide Sales and proportion of franchised shops.")
plt.scatter(x=uswide["U.S. Systemwide Sales (Millions - U.S Dollars)"], y = uswide["Proportion of franchise"])
plt.ylabel("Proportion of franchised shops")
plt.xlabel("U.S. Systemwide Sales (Millions - U.S Dollars)")
```

```python
avg = fastfood.sort_values("Average Sales per Unit (Thousands - U.S Dollars)", ascending = False)
```

```python
plt.figure(figsize = (16,6))
plt.title("Average sales per unit", fontsize = 16)
plt.xticks(rotation = 90)
plt.bar(avg.index, avg["Average Sales per Unit (Thousands - U.S Dollars)"], color = "#74b9ff", label = "Average Sales per unit")
plt.legend()
```

```python
avg[["Average Sales per Unit (Thousands - U.S Dollars)", "Proportion of franchise"]].tail(5)
```

```python
plt.figure(figsize=(4,4))
plt.title("Relationship between Sales and proportion of franchised shops")
plt.scatter(x=avg["Average Sales per Unit (Thousands - U.S Dollars)"], y = avg["Proportion of franchise"])
plt.ylabel("Proportion of franchised shops")
plt.xlabel("Average Sales per unit")
```
